```python
def sigmoid(z):
    val = 1 / (1 + np.exp(-z))
    return val
```

```python
i = np.arange(-5.0, 5, 0.25)
```

```python
i
```

```python
s = sigmoid(i)
```

#### Bar plot

```python
studs = [67,89,120,189,163,69]
courses = ['C','C++','Java','Python','PHP','JS']
```

```python
plt.title('Total Enrollment')
plt.xlabel('Courses')
plt.ylabel('Total Students')
plt.grid()
plt.bar(courses, studs, color = 'magenta')
```

```python
plt.title('Total Enrollment')
plt.ylabel('Courses')
plt.xlabel('Total Students')
plt.grid()
plt.barh(courses, studs, color = 'magenta')
```

```python
std = pd.read_csv('student.csv')
```

```python
plt.figure(figsize=(16,9))
plt.title('Result')
plt.ylabel('Stuent Name')
plt.xlabel('Marks')
plt.grid()
plt.barh(std['name'],std['marks'],
         color = 'brown')
```

#### Pie chart

```python
plt.figure(figsize=(10,10))
plt.pie(studs, labels=courses, 
        autopct='%2.2f%%', shadow=True);
```
